**Proyecto de pruebas A/B**

Objetivos del estudio:

-Objetivo principal: 
evaluar si el cambio asociado al sistema de recomendaciones mejorado (grupo B) incrementa la conversión del embudo product_page → product_cart → purchase dentro de los 14 días posteriores al registro, comparado con el grupo control A.

-Objetivos específicos:

Verificar que la prueba se realizó correctamente (audiencia UE, periodo, asignación única a grupo, tamaño muestral).

Analizar el comportamiento del embudo por grupo y detectar anomalías (eventos desbalanceados, usuarios en ambos grupos, distribución diaria irregular, influencia de campañas de marketing).

Estimar conversiones por etapa y medir el uplift del grupo B.

Confirmar estadísticamente las diferencias mediante prueba z para proporciones por etapa.

-Items a revisar:

Tipos de datos (fechas a datetime, columnas categóricas).

Datos nulos/duplicados por tabla (especialmente en participants y events).

Usuarios en ambos grupos (grave: invalida A/B).

Usuarios en otras pruebas además de recommender_system_test (contaminación).

Región UE y solo usuarios registrados en 2020-12-07 a 2020-12-21.

Ventana de análisis: eventos solo hasta 14 días desde first_date (no usar todo el histórico hasta 2021-01-01 si un usuario se registró tarde).

Balance de eventos por usuario y distribución por día (picos raros pueden indicar sesgo por marketing).

```python
import pandas as pd
import numpy as np
from math import sqrt
from scipy import stats

# =========================
# 0) Configuracion y carga
# =========================

PATH_MARKETING = "/datasets/ab_project_marketing_events_us.csv"
PATH_USERS     = "/datasets/final_ab_new_users_upd_us.csv"
PATH_EVENTS    = "/datasets/final_ab_events_upd_us.csv"
PATH_PART      = "/datasets/final_ab_participants_upd_us.csv"

marketing = pd.read_csv(PATH_MARKETING)
users = pd.read_csv(PATH_USERS)
events = pd.read_csv(PATH_EVENTS)
participants = pd.read_csv(PATH_PART)

print("Shapes:")
print("marketing:", marketing.shape)
print("users:", users.shape)
print("events:", events.shape)
print("participants:", participants.shape)

```

```python
# =========================
# 1. Exploración inicial de los datos (EDA básico)
# =========================

datasets = {
    "marketing": marketing,
    "users": users,
    "events": events,
    "participants": participants
}

for name, df in datasets.items():
    print(f"\n{'='*40}")
    print(f"DATASET: {name}")
    print(f"{'='*40}")

    print("\nPrimeras filas:")
    print(df.head())

    print("\nInformación general:")
    df.info()

    print("\nValores nulos:")
    print(df.isna().sum())

    print("\nDuplicados totales:", df.duplicated().sum())
```

La exploración inicial de los datos muestra que los cuatro conjuntos presentan una estructura consistente y adecuada para el análisis. El dataset marketing contiene 14 campañas publicitarias sin valores nulos ni duplicados, aunque las fechas se encuentran inicialmente como texto y requieren conversión a formato datetime. El dataset users incluye 58 703 registros de usuarios con información completa sobre fecha de registro, región y dispositivo, sin valores faltantes ni duplicados. El dataset events, el más voluminoso, registra más de 423 000 eventos de usuarios; si bien no presenta valores nulos en identificadores ni fechas, la columna details contiene una alta proporción de valores faltantes, lo cual es consistente con el hecho de que solo los eventos de compra incluyen información monetaria. Finalmente, el dataset participants cuenta con 14 525 usuarios correctamente asignados a grupos A o B de la prueba recommender_system_test, sin valores nulos ni duplicados, lo que indica una asignación clara y consistente para el análisis de la prueba A/B.

**Analisis exploratorio de datos:**

```python

import matplotlib.pyplot as plt

# =========================================================
# 2. CONVERSIÓN DE TIPOS
# =========================================================

marketing['start_dt'] = pd.to_datetime(marketing['start_dt'])
marketing['finish_dt'] = pd.to_datetime(marketing['finish_dt'])

users['first_date'] = pd.to_datetime(users['first_date'])
events['event_dt'] = pd.to_datetime(events['event_dt'])

participants['group'] = participants['group'].astype(str).str.strip()

# =========================================================
# 3. FILTRADO DE LA PRUEBA Y VALIDACIÓN DE PARTICIPANTES
# =========================================================

TEST_NAME = "recommender_system_test"

# Solo participantes de la prueba
part_test = participants[participants['ab_test'] == TEST_NAME].copy()

# Detectar usuarios en ambos grupos
users_multi_group = (
    part_test.groupby('user_id')['group']
    .nunique()
    .reset_index()
)
users_multi_group = users_multi_group[users_multi_group['group'] > 1]

# Excluir usuarios contaminados
part_test = part_test[
    ~part_test['user_id'].isin(users_multi_group['user_id'])
]

# =========================================================
# 4. CREACIÓN DE DATA (PARTICIPANTES VÁLIDOS)
# =========================================================

data = part_test.merge(users, on='user_id', how='left')

# Filtros del experimento
data = data[
    (data['first_date'] >= '2020-12-07') &
    (data['first_date'] <= '2020-12-21') &
    (data['region'] == 'EU')
].copy()

print("Participantes finales por grupo:")
print(data['group'].value_counts())

# =========================================================
# 5. CREACIÓN DE EVENTS_U_14 (EVENTOS EN VENTANA DE 14 DÍAS)
# =========================================================

events_u = events.merge(
    data[['user_id', 'group', 'first_date']],
    on='user_id',
    how='inner'
)

events_u['window_end'] = events_u['first_date'] + pd.Timedelta(days=14)

events_u_14 = events_u[
    (events_u['event_dt'] >= events_u['first_date']) &
    (events_u['event_dt'] < events_u['window_end'])
].copy()

print("\nEventos dentro de la ventana de 14 días:")
print(events_u_14.shape)

# =========================================================
# 6. EDA – CONVERSIÓN EN LAS ETAPAS DEL EMBUDO
# =========================================================

funnel_steps = ['product_page', 'product_cart', 'purchase']

user_funnel = (
    events_u_14[events_u_14['event_name'].isin(funnel_steps)]
    .pivot_table(
        index=['user_id', 'group'],
        columns='event_name',
        values='event_dt',
        aggfunc='min'
    )
    .reset_index()
)

for step in funnel_steps:
    user_funnel[step] = user_funnel[step].notna().astype(int)

base_users = data[['user_id', 'group']].drop_duplicates()

user_funnel = base_users.merge(
    user_funnel[['user_id','group'] + funnel_steps],
    on=['user_id','group'],
    how='left'
).fillna(0)

funnel_counts = user_funnel.groupby('group')[funnel_steps].sum()
funnel_counts['n_users'] = user_funnel.groupby('group')['user_id'].nunique()

print("\n=== USUARIOS POR ETAPA DEL EMBUDO ===")
print(funnel_counts)

funnel_conversion = funnel_counts.copy()
for step in funnel_steps:
    funnel_conversion[step] = funnel_conversion[step] / funnel_conversion['n_users']

print("\n=== CONVERSIÓN ACUMULADA POR ETAPA ===")
print(funnel_conversion)

# =========================================================
# 7. EDA – EVENTOS POR USUARIO
# =========================================================

events_per_user = (
    events_u_14
    .groupby(['group','user_id'])
    .size()
    .reset_index(name='events_count')
)

print("\n=== EVENTOS POR USUARIO (DESCRIPTIVOS) ===")
print(events_per_user.groupby('group')['events_count'].describe())

plt.boxplot(
    [
        events_per_user[events_per_user['group'] == 'A']['events_count'],
        events_per_user[events_per_user['group'] == 'B']['events_count']
    ],
    labels=['Grupo A', 'Grupo B'],
    showfliers=True
)

plt.title('Distribución del número de eventos por usuario')
plt.ylabel('Eventos por usuario')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# =========================================================
# 8. EDA – USUARIOS EN AMBAS MUESTRAS
# =========================================================

users_in_both = (
    data.groupby('user_id')['group']
    .nunique()
    .reset_index()
)
users_in_both = users_in_both[users_in_both['group'] > 1]

print("\nUsuarios presentes en ambos grupos:", len(users_in_both))

# =========================================================
# 9. EDA – DISTRIBUCIÓN DIARIA DE EVENTOS
# =========================================================

events_u_14['event_date'] = events_u_14['event_dt'].dt.date

daily_events = (
    events_u_14
    .groupby(['group','event_date'])
    .size()
    .reset_index(name='events_count')
)

plt.figure(figsize=(10,5))
for grp in ['A', 'B']:
    subset = daily_events[daily_events['group'] == grp]
    plt.plot(subset['event_date'], subset['events_count'], label=f'Grupo {grp}')

plt.title('Distribución diaria del número de eventos')
plt.xlabel('Fecha')
plt.ylabel('Número de eventos')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# =========================================================
# 10. PECULIARIDADES ANTES DE LA PRUEBA A/B
# =========================================================

print("\n=== PECULIARIDADES Y CONSIDERACIONES ===")
print("• Solo usuarios UE incluidos.")
print("• Ventana de análisis limitada a 14 días.")
print("• No hay usuarios en ambos grupos.")
print("• Actividad por usuario comparable entre grupos.")
print("• Campañas de marketing activas pueden influir en resultados.")
print("• Valores nulos en 'details' son esperados (solo compras tienen monto).")

```

El análisis exploratorio confirma que la prueba A/B se ejecutó sobre muestras comparables y técnicamente válidas, aunque con tamaños desbalanceados entre los grupos (2 604 usuarios en el grupo A frente a 877 en el grupo B), lo cual debe considerarse al interpretar los resultados. En el embudo de conversión product_page → product_cart → purchase, el grupo A presenta mayores tasas de conversión acumulada en todas las etapas: aproximadamente 64.7% de los usuarios alcanzan la vista de producto frente al 56.2% en el grupo B, mientras que la conversión a compra es de 32.0% en A y 28.4% en B. Esto sugiere que, a nivel descriptivo, el nuevo sistema de recomendaciones no muestra una mejora evidente en el comportamiento de los usuarios.

La distribución del número de eventos por usuario es similar entre ambos grupos, con medianas cercanas (6 eventos en A y 4 en B) y presencia de valores atípicos en ambos casos, lo que indica que no existe un sesgo fuerte de sobreactividad en una sola muestra. Asimismo, no se detectaron usuarios presentes en ambos grupos, preservando el supuesto de independencia requerido para una prueba A/B. La distribución diaria de eventos muestra patrones temporales consistentes entre grupos, con picos de actividad alineados en fechas similares, lo que sugiere que factores externos —como campañas de marketing activas durante el periodo— afectaron de manera comparable a ambas muestras. En conjunto, no se identifican anomalías críticas que invaliden el experimento, por lo que los datos se consideran adecuados para proceder con la evaluación estadística formal de la prueba A/B.

**PRUEBA Z PARA DIFERENCIA DE PROPORCIONES**

```python
# ===============================
# Preparación de datos
# ===============================

A = funnel_counts.loc['A']
B = funnel_counts.loc['B']

n_A = int(A['n_users'])
n_B = int(B['n_users'])

steps = ['product_page', 'product_cart', 'purchase']

results = []

alpha = 0.05

# ===============================
# Prueba Z por etapa del embudo
# Hipótesis:
# H0: p_B <= p_A
# H1: p_B > p_A  (mejora esperada del nuevo sistema)
# ===============================

for step in steps:
    success_A = int(A[step])
    success_B = int(B[step])

    p_A = success_A / n_A
    p_B = success_B / n_B

    # Proporción combinada
    p_pool = (success_A + success_B) / (n_A + n_B)

    # Error estándar
    se = sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))

    # Estadístico Z
    z = (p_B - p_A) / se if se > 0 else np.nan

    # p-value unilateral (B > A)
    p_value = 1 - stats.norm.cdf(z)

    results.append([
        step,
        p_A,
        p_B,
        p_B - p_A,
        z,
        p_value,
        p_value < alpha
    ])

# ===============================
# Resultados finales
# ===============================

z_test_results = pd.DataFrame(
    results,
    columns=[
        'Etapa',
        'Conversión A',
        'Conversión B',
        'Diferencia (B - A)',
        'Z',
        'p_value',
        'Rechaza H0 (α=0.05)'
    ]
)

print("=== RESULTADOS PRUEBA Z (B > A) ===")
print(z_test_results)
```

**Interpretación de resultados de la prueba Z:**

Los resultados de la prueba z para la diferencia de proporciones muestran que el grupo B presenta tasas de conversión inferiores al grupo A en todas las etapas del embudo. En la etapa product_page, la conversión del grupo B es aproximadamente 8.5 puntos porcentuales menor que la del grupo A, diferencia que resulta estadísticamente significativa en sentido negativo (Z = −4.50), lo que indica un desempeño claramente inferior del grupo experimental. Sin embargo, dado que la hipótesis alternativa evaluada asume una mejora del grupo B sobre el grupo A, el valor p obtenido es cercano a 1, por lo que no se rechaza la hipótesis nula.

En las etapas product_cart y purchase, las diferencias entre grupos también son negativas (−2.2 y −3.6 puntos porcentuales, respectivamente), aunque de menor magnitud. En estos casos, los valores del estadístico Z (−1.24 y −1.99) y los valores p asociados confirman que no existe evidencia estadísticamente significativa de que el nuevo sistema de recomendaciones mejore la conversión en ninguna fase del embudo. En consecuencia, bajo el nivel de significancia del 5%, la hipótesis nula no puede ser rechazada en ninguna de las etapas analizadas.

**Conclusión**

La evaluación estadística de la prueba A/B confirma que el nuevo sistema de recomendaciones (grupo B) no mejora el rendimiento del embudo de conversión en comparación con el grupo control. Por el contrario, los resultados muestran consistentemente tasas de conversión más bajas para el grupo experimental en todas las etapas del proceso de compra. Dado que ninguna de las diferencias observadas resulta estadísticamente significativa a favor del grupo B, no se recomienda la implementación del nuevo sistema en su estado actual. Se sugiere revisar el diseño del sistema de recomendaciones, analizar posibles efectos derivados del desbalance en el tamaño de las muestras y considerar una nueva iteración del experimento antes de tomar decisiones de despliegue.
